```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with np.load("../Dataset/NoReg_Extra_6.npz") as data:
    x1 = data['images']
    
with np.load("../Dataset/mnist.npz") as data:
    x2 = data['x_train']

x1 = x1.astype("float32") / 255.0
x2 = x2.astype("float32") / 255.0

x1 = x1.reshape(-1, 28 * 28)
x1 = x1.mean(axis=1)

x2 = x2.reshape(-1, 28 * 28)
x2 = x2.mean(axis=1)
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(x1, label="NoReg_Extra_6", fill=True, alpha=0.5)
sns.kdeplot(x2, label="MNIST", fill=True, alpha=0.5)
plt.xlabel("Pixel Intensity")
plt.ylabel("Density")
plt.title("KDE Comparison of Pixel Intensity Distributions")
plt.legend()
plt.show()
```
